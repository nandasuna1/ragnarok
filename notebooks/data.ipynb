```python
import pandas as pd

# Caminho para os dados
path = "../data/raw/download.csv"
# Leitura do CSV
df = pd.read_csv(path)

# Exibir as primeiras linhas
df.head()
```

```python
# Ver as colunas
df.columns

# InformaÃ§Ãµes gerais
df.info()

# Ver valores nulos
df.isnull().sum()

# Amostra de registros completos
df.sample(5)
```

```python
# Preencher valores ausentes com string vazia para evitar erro
df.fillna(value={col: "" for col in df.select_dtypes(include="object").columns}, inplace=True)

# Gerar os documentos concatenando as informaÃ§Ãµes mais Ãºteis
def construir_documento(row):
    return (
        f"Date: {row['EventDate']}. Local: {row['City']}, {row['State']}, {row['Country']}. "
        f"Injuries: {row['HighestInjuryLevel']} (Fatais: {row['FatalInjuryCount']}, "
        f"Serious: {row['SeriousInjuryCount']}, Leves: {row['MinorInjuryCount']}). "
        f"Probable Cause: {row['ProbableCause']}. CondiÃ§Ã£o climÃ¡tica: {row['WeatherCondition']}. "
        f"AirCraft Damage: {row['AirCraftDamage']}. PropÃ³sito do voo: {row['PurposeOfFlight']}. "
        f"Operator: {row['Operator']}."
    )

# Criar nova coluna com os documentos
df["document"] = df.apply(construir_documento, axis=1)

# Ver exemplo
df["document"].iloc[0]
```

```python
df.to_csv("../data/processed/clean_data.csv", index=False)
```
